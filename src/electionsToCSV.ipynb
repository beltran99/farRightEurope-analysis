```python
countries = ["FR", "AT", "IE", "UK", "BG", "UA", "PL", "SK", "SI", "TR"]
headers = {
    "FR":['Region','Year','Party','FirstRoundVotes','FirstRoundPercentage','FirstRoundSeats','SecondRoundVotes','SecondRoundPercentage','SecondRoundSeats','TotalSeats'],
    "AT":['Region','Year','Party','Votes','VotesPercentage','TotalSeats'],
    "IE":['Region','Year','Party','Votes','VotesPercentage','TotalSeats'],
    "UK":['Region','Year','Party','Votes','VotesPercentage','TotalSeats'],
    "BG":['Region','Year','Party','Votes','VotesPercentage','TotalSeats'],
    "UA":['Region','Year','Party','Votes','VotesPercentage','TotalSeats'],
    "PL":['Region','Year','Party','Votes','VotesPercentage','TotalSeats'],
    "SK":['Region','Year','Party','Votes','VotesPercentage','TotalSeats'],
    "SI":['Region','Year','Party','Votes','VotesPercentage','TotalSeats'],
    "TR":['Region','Year','Party','Votes','VotesPercentage','TotalSeats']
           }
years = {
    "FR":[2002, 2007, 2012, 2017],
    "AT":[2002, 2006, 2008, 2013, 2017],
    "IE":[2002, 2007, 2011, 2016, 2020],
    "UK":[2001, 2005, 2010, 2015, 2017, 2019],
    "BG":[2001, 2005, 2009, 2013],
    "UA":[2006, 2007, 2012, 2014, 2019],
    "PL":[2001, 2005, 2007, 2011, 2015],
    "SK":[2002, 2006, 2010, 2012, 2016],
    "SI":[2004, 2008, 2011, 2014, 2018],
    "TR":[2002, 2007, 2011, 2018]
    }

url1 = 'http://electionresources.org/'
urls = {
    "FR":'/deputies.php?election=',
    "AT":'/nationalrat.php?election=',
    "IE":'/house.php?election=',
    "UK":'/house.php?election=',
    "BG":'/assembly.php?election=',
    "UA":'/deputies.php?election=',
    "PL":'/sejm.php?election=',
    "SK":'/nrsr.php?election=',
    "SI":'/assembly.php?election=',
    "TR":'/assembly.php?election='
}
```

```python
for c in countries:
  print(c)
  election_years = years[c]

  for y in election_years:
    print(y,'\n')

    url2 = urls[c]
    url = url1 + c.lower() + url2 + str(y)

    try:
      data = urllib.request.urlopen(url)
    except Exception as e:
      url = url1 + c.lower() + '/' + str(y) + '.php'
      data = urllib.request.urlopen(url)
    page = requests.get(url=url, timeout=2)
    soup = BeautifulSoup(page.content, 'html.parser')
    data1 = soup.find_all(bgcolor="#EFEFEF")
    data2 = soup.find_all(bgcolor="#FFFFFF")
    data = data1 + data2

    csv_name = "files/" + c + "_" + str(y) + ".csv"
    with open(csv_name, 'w', encoding='UTF8') as fwrite:

      writer = csv.writer(fwrite)
      h = headers[c]
      writer.writerow(h)

      numerical = len(h) - 3

      for d in data:
        line = [c,y]
        var = unidecode.unidecode(d.get_text()).split()
        name = var[:-numerical]
        numbers = var[-numerical:]
        name = " ".join(name)
        print(name)
        line.append(name)
        for n in numbers:
          if n == "-":
            n = .0
          else:
            n = float(n.replace(',',''))
          print(n)
          line.append(n)
        print('\n')
        writer.writerow(line)
      print('#####################################\n')
```
